```python
# Importamos las librerias necesarias para el uso de los datos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Procedemos a leer los datos descargados de la pagina del Observatorio Valentia
# Saltamos las primeras 6 lineas ya que no hay datos ahi
dfD = pd.read_csv('ValentiaObsDec.txt', skiprows=4, sep='\s+')
dfJ = pd.read_csv('ValentiaObsJun.txt', skiprows=4, sep='\s+')

#Una vez leidos, nombramos las columnas
dfD.columns = ['PRES','HGHT', 'TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
dfJ.columns = ['PRES','HGHT', 'TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
dfD = dfD.drop(dfD.index[0])
dfJ = dfJ.drop(dfJ.index[0])
```

```python
dfD
```

```python
#Eliminamos las ultimas columnas, que no forman parte de los datos necesarios.
dfD=dfD[:-26]
dfJ=dfJ[:-26]
```

```python
dfJ.dtypes
```

```python
#Transformamos los objetos a numero
columnas = ['PRES','HGHT', 'TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
dfJ[columnas]= dfJ[columnas].apply(pd.to_numeric,errors='coerce', axis=1)
dfD[columnas]= dfD[columnas].apply(pd.to_numeric,errors='coerce', axis=1)
```

```python
dfJ.dtypes
```

```python
#Grafica de Presion contra Altura en el mes de Diciembre
ax = dfD.plot(x='PRES', y='HGHT', color='black', alpha=1, linewidth=3)
ax.set_xlim((0, 1050))
ax.set_title("Grafica Presión con respecto a Altura. Diciembre.")
ax.set_xlabel("Presión (hPa)")
ax.set_ylabel("Altura (m)")
ax.grid(color='black', linestyle='-', linewidth=0.3)

#Con esto le damos color a la grafica
ax.set_facecolor("cornflowerblue")

plt.show()
```

```python
#Grafica de Presion contra Altura en el mes de Junio
ax = dfJ.plot(x='PRES', y='HGHT', color='black', alpha=1, linewidth=3)
ax.set_xlim((0, 1050))
ax.set_title("Grafica Presión con respecto a Altura. Junio.")
ax.set_xlabel("Presión (hPa)")
ax.set_ylabel("Altura (m)")

# Color de fondo
ax.set_facecolor("lightsalmon")
ax.grid(color='black', linestyle='-', linewidth=0.3)
plt.show()
```

```python
#Grafica de Temperatura contra Altura en el mes de Diciembre
ax = dfD.plot(x='TEMP', y='HGHT', color='black', alpha=1, linewidth=3)
ax.set_xlim((-80, 20))
ax.set_title("Grafica Temperatura con respecto a Altura. Diciembre.")
ax.set_xlabel("Temperatura (°C)")
ax.set_ylabel("Altura (m)")
ax.grid(color='black', linestyle='-', linewidth=0.3)

#Con esto le damos color a la grafica
ax.set_facecolor("cornflowerblue")

plt.show()
```

```python
#Grafica de Temperatura contra Altura en el mes de Junio
ax = dfJ.plot(x='TEMP', y='HGHT', color='black', alpha=1, linewidth=3)
ax.set_xlim((-80, 20))
ax.set_title("Grafica Temperatura con respecto a Altura. Junio.")
ax.set_xlabel("Temperatura (°C)")
ax.set_ylabel("Altura (m)")

# Color de fondo
ax.set_facecolor("lightsalmon")
ax.grid(color='black', linestyle='-', linewidth=0.3)
plt.show()
```

```python
#Grafica de Temperaturas contra Altura en el mes de Diciembre
x1=dfD['TEMP']
x2=dfD['DWPT']
y=dfD['HGHT']

plt.plot(x1, y, "navy", label='TEMP')
plt.plot(x2, y, "cornflowerblue", label='DWPT')
plt.legend(loc='upper right')
plt.title("Grafica Temperatura y Temperatura de rocío con respecto a Altura. Diciembre.")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Altura (m)")

plt.grid(True)

plt.show()
```

```python
#Grafica de Temperaturas contra Altura en el mes de Junio
x1=dfJ['TEMP']
x2=dfJ['DWPT']
y=dfJ['HGHT']

plt.plot(x1, y, "red", label='TEMP')
plt.plot(x2, y, "lightsalmon", label='DWPT')
plt.legend(loc='upper right')
plt.title("Grafica Temperatura y Temperatura de rocío con respecto a Altura. Diciembre.")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Altura (m)")

plt.grid(True)

plt.show()
```
